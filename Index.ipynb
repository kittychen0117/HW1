```python
import csv
from io import BytesIO
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://www.census.gov/manufacturing/m3/historical_data/index.html"
r = requests.get(url)
r.encoding = 'utf8'
soup = BeautifulSoup(r.text, 'html.parser')
head = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36'}

i = 0
df1 = pd.DataFrame()

for links in soup.find_all('a', string="Excel"):
    if (i<6):
        link = "https://www.census.gov" + links.get('href')
    else:
        link = "https://www.census.gov/manufacturing/m3/historical_data/" + links.get('href')
    #print (link)
    i = i+1
    if (i==12):
        break
    download = requests.get(link, headers=head)
    #print (download)
    df = pd.read_excel(BytesIO(download.content))
    df.columns = ['Name','Year','Value','','','','','','','','','','','']
    print (links.get('title'))
    print(df.head(20))
    print("\n--------------------------------------------------------------------------------------------------\n\n")
```
